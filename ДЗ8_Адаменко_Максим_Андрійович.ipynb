```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma, normaltest

shape = 0.3
scale = 1.1
x = gamma.rvs(shape, scale=scale, size=1000)


plt.hist(x, bins=30, density=True, color='skyblue', edgecolor='black')
plt.title('Гістограма розподілу x')
plt.xlabel('Значення')
plt.ylabel('Ймовірність')
plt.grid(True)
plt.show()


def simulate_stock_price(t):
    return np.sum(gamma.rvs(shape, scale=scale, size=t))

t_values = np.arange(1, 61, 2)
for t in t_values:
    prices = [simulate_stock_price(t) for _ in range(100)]  # Симуляція 100 разів
    plt.hist(prices, bins=30, density=True, color='skyblue', edgecolor='black')
    plt.title(f'Гістограма цін акцій для t = {t}')
    plt.xlabel('Ціна акцій')
    plt.ylabel('Ймовірність')
    plt.grid(True)
    plt.show()


    stat, p = normaltest(prices)
    if p < 0.05:
        print(f'Розподіл для t = {t} не є нормальним (p-value = {p})')
    else:
        print(f'Розподіл для t = {t} є нормальним (p-value = {p})')
```

```python
import pandas as pd

url =  'https://docs.google.com/spreadsheets/d/18WCpPS96Tb3cB0FCsIA92PEhcmBkp08sjYhS9DsQfJE/edit?usp=sharing'
url = url[:url.find('/edit')] + '/export?format=csv'

df = pd.read_csv(url)

stats = df.describe()

print("Середнє значення:")
print(stats.loc['mean'])
print("\nДисперсія:")
print(stats.loc['std'] ** 2)
print("\nСтандартне відхилення:")
print(stats.loc['std'])


df.hist(bins=20, figsize=(10, 8))
plt.tight_layout()
plt.show()


from scipy.stats import normaltest

for column in df.columns[:-1]:
    print(f"\n{column}:")
    stat, p = normaltest(df[column])
    if p < 0.05:
        print(f'Розподіл не є нормальним (p-value = {p})')
    else:
        print(f'Розподіл є нормальним (p-value = {p})')

    correlation = df[column].corr(df['Product_Sold'])
    print(f"Кореляція з Product_Sold: {correlation}")
```
