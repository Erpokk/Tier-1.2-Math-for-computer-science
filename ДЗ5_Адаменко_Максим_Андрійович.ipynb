Виконай диференціювання функції та знайди значення похідних у точках
$x=1$ та $x=-\frac{1}{2}$. \
Розв’язати програмно та аналітично.\
1)$\frac{x^3}{3} +\frac{x^2}{2} -2x $\
Спочатку треба знайти похідну від нашого виразу \
$f'(x) = x^2 + x -2$\
Далі підставляємо значення в цю похідну:\
Для $x=1$\
$f'(1) = 1^2 + 1 -2 = 1+1-2=0 $\
Для $x=-\frac{1}{2}$\
$f'(-\frac{1}{2}) = (-\frac{1}{2})^2 -\frac{1}{2} -2 =-\frac{9}{4} $


```python
import sympy as sp


x = sp.symbols('x')
f = sp.sqrt(x**3/3 + x**2/2 - 2*x)

# Знаходження похідної
f_prime = sp.diff(f, x)

x_values = [1, -1/2]
derivatives = [f_prime.subs(x, val).evalf() for val in x_values]

print("Значення похідної у точках:")
for x_val, der in zip(x_values, derivatives):
    print(f"x = {x_val}: {der}")
```

Виконай диференціювання функції та знайди значення похідних у точках
$x=1$ та $x=-\frac{1}{2}$. \
Розв’язати програмно та аналітично.\
2)$\sqrt{x^2+1}$\
Cпочатку перепишемо наш вираз як $(x^2+1)^\frac{1}{2}$ та скориставшись правилом ланцюга знаходимо:
$f'(х) = \frac{1}{2}*(x^2+1)^{-\frac{1}{2}} *2x = \frac{x}{\sqrt{x^2+1}}$ \
Підставляємо значення в цю похідну:\
Для $x=1$\
$f'(1) = \frac{1}{\sqrt{1^1+1}} = \frac{1}{\sqrt{2}} $\
Для $x=-\frac{1}{2}$\
$f'(-\frac{1}{2}) = \frac{-\frac{1}{2}}{\sqrt{(-\frac{1}{2})^2+1}} = \frac{-\frac{1}{2}}{\sqrt{\frac{5}{4}}} = -\frac{1}{2} * \frac{\sqrt4}{\sqrt5} = -\frac{1}{\sqrt5}  $


```python
import sympy as sp

# Визначення символьної змінної та функції
x = sp.symbols('x')
f = sp.sqrt(x**2 + 1)

# Знаходження похідної
f_prime = sp.diff(f, x)

# Обчислення значень похідних у точках x=1 та x=-1/2
x_values = [1, -1/2]
derivatives = [f_prime.subs(x, val).evalf() for val in x_values]

print("Значення похідної у точках:")
for x_val, der in zip(x_values, derivatives):
    print(f"x = {x_val}: {der}")
```

Виконай диференціювання функції та знайди значення похідних у точках
$x=1$ та $x=-\frac{1}{2}$. \
Розв’язати програмно та аналітично.\
3)$\frac{1}{\sqrt{x^2+1}}$\
Cпочатку перепишемо наш вираз як $\frac{1}{(x^2+1)^\frac{1}{2}}$ та скориставшись правилом ланцюга знаходимо:
$f'(х) = \frac{1}{2}*(x^2+1)^{-\frac{3}{2}} *2x = \frac{x}{(x^2+1)^{\frac{3}{2}}}$ \
Підставляємо значення в цю похідну:\
Для $x=1$\
$f'(1) = -\frac{1}{(1^2+1)^{\frac{3}{2}}} = -\frac{1}{\sqrt2}$\
Для $x=-\frac{1}{2}$\
$f'(-\frac{1}{2}) = -\frac{-\frac{1}{2}}{((-\frac{1}{2})^2+1)^{\frac{3}{2}}} = -\frac{-\frac{1}{2}}{\frac{5}{4}^{\frac{3}{2}}} = -\frac{-\frac{1}{2}}{\frac{\sqrt125}{8}} = \frac{1}{2} * \frac{4}{\sqrt{125}} = \frac{4}{\sqrt{125}}$

```python
import sympy as sp

# Оголошуємо символьну змінну x
x = sp.symbols('x')

# Оголошуємо функцію
f = 1 / sp.sqrt(x**2 + 1)

# Знаходимо похідну функції
f_prime = sp.diff(f, x)

# Задаємо точки, у яких потрібно знайти значення похідної
x_values = [1, -1/2]

# Обчислюємо значення похідної у заданих точках
derivatives = [f_prime.subs(x, val).evalf() for val in x_values]

# Виводимо результати
print("Значення похідної у точках:")
for x_val, der in zip(x_values, derivatives):
    print(f"x = {x_val}: {der}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Оголошуємо символьну змінну x
x = sp.symbols('x')

# Оголошуємо функцію
f = -3 * x**2 + 30 * x

# Знаходимо похідну функції
f_prime = sp.diff(f, x)

# Задаємо функцію для обчислення значень
f_lambda = sp.lambdify(x, f, 'numpy')
f_prime_lambda = sp.lambdify(x, f_prime, 'numpy')

# Задаємо діапазон значень x
x_values = np.linspace(0, 12, 200)
y_values = f_lambda(x_values)

# Знаходимо точку максимуму по х та у
critical_points = sp.solve(f_prime, x)
max_x = max(critical_points)
max_y = f_lambda(max_x)

# Виводимо графіки
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='Функція f(x)', color='blue')
plt.plot(x_values, f_prime_lambda(x_values), label="Похідна f'(x)", linestyle='--', color='red')
plt.scatter(max_x, max_y, color='green', label=f'Точка максимуму (x={max_x}, y={max_y})')
plt.title('Графік функції та її похідної')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

```
